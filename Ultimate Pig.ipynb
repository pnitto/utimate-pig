```python
import random

class Player:
    
    def __init__(self):
        self.point_count = 0
        self.round_count = 0
    
    
    def hold_or_stay(self, value_of_die):
        return random.choice(["Stay","Roll"])
    
        
class Game:
    
    def __init__(self, player):
        self.player = player
        
    def roll_die(self):
        roll = random.randint(1,6)
        return roll
    
    def simulate(self):
        print("TOTAL POINTS: ", self.player.point_count)
        die_value = self.roll_die()
        if die_value == 1:
            print("You rolled a 1. You get zero points.")
            self.player.round_count = 0
            return self.player.round_count
        player_choice = self.player.hold_or_stay(die_value)
        print("ROLL: ", die_value)
        print("CHOICE: ", player_choice)
        if player_choice == "Stay":
            self.add_points_to_player(die_value)
            self.total_points_of_round()
            print("TOTAL POINTS END OF GAME: ", self.player.point_count)
        elif player_choice == "Roll":
            self.add_points_to_player(die_value)
            print("ROUND COUNT: ", self.player.round_count)
            self.simulate()
    
    def add_points_to_player(self, die_value):
        self.player.round_count += die_value
        
    def total_points_of_round(self):
        self.player.point_count += self.player.round_count
        self.round_count = 0
        
          
player_1 = Player()

game = Game(player_1)
game.simulate()
print(game.player.point_count)
        
```
